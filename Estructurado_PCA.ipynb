Ahora, para ver la viabilidad de realizar componentes principales, necesitamos que algunos datos esten correlacionados entre si, para esto emplearemos la prueba de esfericidad de Barlett en donde queremos probar si


$$
H_0: \Sigma = I
$$

$$
H_1: \Sigma \neq I
$$


Para esto emplearemos la prueba contenida en la función cortest.barlett de el paquete psych


```r
install.packages("psych")
library(psych)
```

```r
R<-cor(ciudadesNum)
```

```r
cortest.bartlett(R,n=50)
```

Note que:
$$
p - valor = 4.17519706806139e-30 < 0.05 = \alpha
$$

por lo que hay suficiente evidencia en la muestra para rechazar $$
H_0: \Sigma = I
$$
Es decir, es viable realizar componentes principales

```r
componentes<-prcomp(ciudadesNum, center=TRUE, scale.=TRUE)
```

```r
summary(componentes)
```

El análisis de componentes principales muestra que los primeros 10 componentes explican más del 90% de la variabilidad total, con el primer componente (PC1) capturando el 20.67% y los primeros 3 componentes explicando conjuntamente el 45.24%. La desviación estándar del PC1 es 2.0343, lo que implica que captura la variabilidad equivalente a aproximadamente 4.14 variables. A medida que avanzamos hacia componentes posteriores, la proporción de varianza explicada disminuye, con componentes como PC8 en adelante explicando menos del 5% cada uno y con desviaciones estándar cada vez menores. Esto sugiere que la mayor parte de la variabilidad está concentrada en los primeros componentes, siendo los más relevantes para capturar la estructura de los datos, y que utilizar más de 10 componentes probablemente no aporte mucha información adicional significativa.

Ahora empleando el criterio de la varianza acumulada, mediante un umbral regido por la ley de los pocos vitales, es decir del 80% hemos decidido quedarnos con **los primeros 9 componentes** los cuales atrapan una proporción de varianza explicada acumulada del **83%**

```r
componentes2<-prcomp(ciudadesNum, center=TRUE, scale.=TRUE, rank=9)
```

```r
print(componentes2)
```

```r
library(reshape2)
library(ggplot2)

cargas <- componentes2$rotation
cargasdf <- as.data.frame(cargas)
cargasdf$medicion <- rownames(cargasdf)
cargasheat <- reshape2::melt(cargasdf)
colnames(cargasheat) <- c("variable", "componente", "carga")
graficar_componentes <- function(df, componentes) {

  df_filtrado <- df[df$componente %in% componentes, ]

  ggplot(df_filtrado, aes(x = componente, y = variable, fill = carga, label = sprintf("%0.2f", round(carga, digits = 2)))) +
    geom_tile() +
    scale_fill_distiller(palette = "RdBu", limits = c(-1, 1)) +
    geom_text(size = 3) +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1, size = 10),
          axis.text.y = element_text(size = 10),
          plot.margin = unit(c(1, 1, 1, 1), "cm"),
          panel.grid = element_blank()) +
    labs(fill = "carga", x = "componente", y = "variable") +
    coord_fixed(ratio = 0.7)
}

graficar_componentes(cargasheat, c("PC1", "PC2", "PC3"))
graficar_componentes(cargasheat, c("PC4", "PC5", "PC6"))
graficar_componentes(cargasheat, c("PC7", "PC8", "PC9"))
```

Ahora para un mejor analisis emplearemos la interpretación de la rotación de los componentes

```r
library(psych)
componentes3<-principal(ciudadesNum,nfactors=9,rotate="varimax")

componentes3$communality
componentes3$loadings
```

```r
for (i in 1:9) {
  assign(paste0("componente", i, "_1"), componentes3$loadings[, i])
}
```

```r
library(reshape2)
library(ggplot2)

for (i in seq(1, 9, by = 3)) {
  componente1 <- componentes3$loadings[, i]
  componente2 <- componentes3$loadings[, i+1]
  componente3 <- componentes3$loadings[, i+2]

  cargasdfr <- as.data.frame(cbind(componente1, componente2, componente3))
  cargasdfr$medicion <- rownames(cargasdfr)

  cargasheatr <- reshape2::melt(cargasdfr)

  # Asignar los nombres correctos para los componentes
  colnames(cargasheatr) <- c("variable", "componente", "carga")
  cargasheatr$componente <- factor(cargasheatr$componente,
                                   levels = c("componente1", "componente2", "componente3"),
                                   labels = c(paste0("componente ", i),
                                              paste0("componente ", i+1),
                                              paste0("componente ", i+2)))

  print(
    ggplot(cargasheatr, aes(x = componente, y = variable, fill = carga, label = sprintf("%0.2f", round(carga, digits = 2)))) +
      geom_tile() +
      scale_fill_distiller(palette = "RdBu", limits = c(-1, 1)) +  # Ajustar los límites de la escala de colores
      geom_text(size = 3) +  # Ajustar el tamaño del texto dentro de las celdas
      theme_minimal() +  # Tema minimalista para el gráfico
      theme(axis.text.x = element_text(angle = 45, hjust = 1, size = 10),  # Rotar las etiquetas del eje x
            axis.text.y = element_text(size = 10)) +  # Ajustar el tamaño de las etiquetas del eje y
      labs(fill = "carga", x = "componente", y = "variable")  # Etiquetas de los ejes y la leyenda
  )
}
```

**Componente 1**:
- **Cargas positivas altas**: Tasa de pobreza, Tasa de desempleo, Tasa de alfabetización
- **Cargas negativas altas**: Indice de contaminación, PIB, Número de universidades

Esta componente parece relacionar tasas sociales como pobreza, desempleo y alfabetización con una menor contaminación, PIB y número de universidades. Podríamos llamarla **"Desafíos socioeconómicos contra desarrollo urbano y educativo"**.

**Componente 2**:
- **Cargas positivas altas**: Temperatura media anual, UTC Offset
- **Cargas negativas altas**: Altitud media, Número de universidades

Esta componente parece asociar ciudades con una mayor temperatura y una menor altitud, menor número de universidades, ademas de horarios cercanos al UTC 0. Podríamos llamarla **"Ciudades costeras"**.

**Componente 3**:
- **Cargas positivas altas**: Población, PIB
- **Cargas negativas altas**: Tasa de homicidios, Tasa de alfabetización

Esta componente parece asociar una mayor población y PIB con una menor tasa de homicidios y alfabetización. Podríamos llamarla **"Ciudades con amplia población y PIB pero con baja alfabetización"**.

**Componente 4**:
- **Cargas positivas altas**: Tasa de crecimiento poblacional, Superficie total.
- **Cargas negativas altas**: Ninguna sobresaliente

Esta componente parece mostrar ciudades un mayor crecimiento poblacional, mayor superficie. Podríamos llamarla **"Crecimiento y expansión territorial de ciudades latinoamericanas"**.

**Componente 5**:
- **Cargas positivas altas**: Índice de contaminación, Índice de calidad del aire
- **Cargas negativas altas**: Ninguna sobresaliente

Esta componente representa las ciudades con mayor contaminación aunque paradojicamente mejor calidad del aire, por lo que podriamos tomarla como un error de muestreo o no. Podríamos llamarla **"Ciudades con mayor contaminación pero mejor calidad del aire"**.

**Componente 6**:
- **Cargas positivas altas**: Densidad de la población
- **Cargas negativas altas**: Esperanza de vida

Esta componente parece asociar una menor esperanza de vida con una mayor densidad de población, lo cual puede ser una caracteristica interesante de la muestra. Podríamos llamarla **"Ciudades con mayor población pero menor esperanza de vida"**.

**Componente 7**:
- **Cargas positivas altas**: Escala Exp. de Contaminación
- **Cargas negativas altas**: Precipitación media anual, UTC Offset

Esta componente parece asociar una mayor tasa de contaminación exponencial con una menor precipitación anual y menor desplazamiento del UTC. Podríamos llamarla **"Ciudades con menor precipitación y mayor contaminación"**.

**Componente 8**:
- **Cargas positivas altas**: Tasa de homicidios, Tasa de criminalidad
- **Cargas negativas altas**: UTC Offset moderada

Esta componente parece ubicar las ciudades de mayores tasas de homicidios y criminalidad anual en aquellas que se encuentran ubicadas en las locaciones con menor desplazamiento del UTC. Podríamos llamarla **"Ciudades con mayor criminalidad"**.

**Componente 9**:
- **Cargas positivas altas**: Precipitación media anual
- **Cargas negativas altas**: Salario promedio

Esta componente parece asociar una mayor precipitación anual con menores salarios, sin embargo hay que recordar que pirmero que todo esta es una tecnica descriptiva y la correlación no implica causalidad. Podríamos llamarla **"Ciudades con alta precipitación y salarios bajos"**.
